```python
# Library Pengolahan Data
import pandas as pd
import numpy as np

# Library Visualisasi Data
import matplotlib.pyplot as plt
import seaborn as sns

# Library Model Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset CCV 5 - Supervised Learning Regression.csv")
```

```python
df.info()
```

```python
# Visualisasi Sebaran Data antara Variabel Dependen Salary dan Variable Independen Years
plt.figure(figsize = (12,5))
sns.pairplot(df, x_vars =['Diabetasol Coklat 1 kg'], y_vars = ['Daily Sales'], height = 5, kind = 'scatter')
plt.xlabel('Diabetasol Coklat 1 kg')
plt.ylabel('Daily Sales')
plt.title('Scatterplot antara Penjualan Harian Diabetasol Coklat 1 kg dengan Total Penjualan Harian')
plt.show()
```

Menghapus Missing Value

```python
df.dropna(subset='Diabetasol Coklat 1 kg', inplace=True)
```

Membagi Data Menjadi Data Latih dan Data Uji

```python
X = df['Diabetasol Coklat 1 kg']
X.head()
```

```python
y = df['Daily Sales']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=100)
```

```python
X_train= X_train.to_numpy()
X_test = X_test.to_numpy()
```

```python
X_train = X_train.reshape(X_train.shape[0],1)
X_test = X_test.reshape(X_test.shape[0], 1)
```

Menyesuaikan Garis Regressi dengan Sebaran Data(fit)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Model Melakukan Prediksi
y_pred = lr.predict(X_test)
```

```python
c = [i for i in range(1, len(y_test)+1, 1)]
plt.plot(c, y_test, color='r', linestyle='-', label='Actual Data')
plt.plot(c, y_pred, color='b', linestyle='-', label='Predict Data')
plt.xlabel('Actual Data')
plt.ylabel('Predict Data')
plt.legend()
plt.title('Perbandingan Data Actual dengan Data Predict')
plt.show()
```

```python
# Metriks Evaluasi Model
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error :", mse)
```

```python
print("Intercept of the Model : ", lr.intercept_)
```

```python
print("Coefficient of the Line : ", lr.coef_)
```
